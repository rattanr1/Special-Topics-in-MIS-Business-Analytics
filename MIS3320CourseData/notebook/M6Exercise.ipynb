```python
#import ebola data set
ebola = pd.read_csv('..\data\country_timeseries.csv')
ebola.head(10)
```

```python
#interpolate (fill or insert) missing value using existing data (use with caution!)
ebola_interpolate = ebola.interpolate()
ebola_interpolate.head(10)
```

```python
#Interpolation uses existing values to fill in missing values. This can compromise the accuracy of data if not used correctly.
```

```python
#If any NA values are present, drop that row.
ebola_drop=ebola.dropna()
ebola_drop.shape
```

```python
#only row 19, thus not a good idea to drop
ebola_drop.shape
```

```python
#Use the thresh parameter in dropna() to keep only rows where a number of valid values is available.
```

```python
#If all values are NA, drop that row
ebola_drop=ebola.dropna(how='all')   #to drope whole row as a null value
ebola_drop.shape

#but 0 rows were dropped, meaning that every single row has value there since retained 122 rows
```

```python
#put a THRESHOLD to keep the rows with 1 valid values
ebola_drop=ebola.dropna(thresh=1)  
ebola_drop.shape
```

```python
#put a THRESHOLD to keep the rows with 4 valid values
ebola_drop=ebola.dropna(thresh=4)  
ebola_drop.shape
```
