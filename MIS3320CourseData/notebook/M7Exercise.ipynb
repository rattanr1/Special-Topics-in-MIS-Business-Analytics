```python
#import pandas, seaborn, and the tips dataset
import pandas as pd
import seaborn as sns
tips = sns.load_dataset ('tips')
tips.head(10)
```

```python
#show the data types in the data set
tips.info()
```

```python
#import the statsmodels library (the formula api) to perform linear regression
#simple linear regression (one predictor)
#predict tip using total_bill

import statsmodels.formula.api as smf
model = smf.ols(formula='tip ~ total_bill', data=tips)
results = model.fit()
results.summary()
```

```python
import pandas as pd
import seaborn as sns
tips = sns.load_dataset('tips')
print(tips.head())
```

```python
#The goal of linear regression is to draw a straight-line relationship between
#a response variable (also known as an outcome or dependent variable) and a
#predictor variable (also known as a feature, covariate, or independent
#variable).

#In our simple linear regression above, we’d like to see how the total_bill relates to or predicts the tip
```

```python
#multiple regression (multiple predictors)
#predict tip using total_bill and size
model = smf.ols(formula='tip ~ total_bill + size' , data=tips)
results = model.fit()
results.summary()
```

```python
#list unique values for sex
tips.sex.unique()

```

```python
#multiple regression with categorical variables
model = smf.ols(formula='tip ~ total_bill + size + sex + smoker + day + time' , data=tips)
results = model.fit()
results.summary()
```
