```python
import matplotlib.pyplot as plt
import seaborn as sns
anscombe = sns.load_dataset('anscombe')
anscombe
                                                #The result shows the anscombe data set
                                                
```

```python
#Subset the Data
```

```python
dataset_1 = anscombe[anscombe['dataset']=='I']
dataset_2 = anscombe[anscombe['dataset']=='II']
dataset_3 = anscombe[anscombe['dataset']=='III']
dataset_4 = anscombe[anscombe['dataset']=='IV']
#This subset the anscombe data set. 
#(The anscombe is intentionally designed so that each data set has the same count, mean, and standard deviation.)

#subset contains set 1-4 of anscombe. Each has same count, mean, and std.
```

```python
#Make a Figure with Axes
```

```python
#use figure( ) and add_subplot( ) for figure and axes
fig = plt.figure(figsize=(10,10))

#add figure title
fig.suptitle('Anscombe Data by Riyan Rattan')      #REPLACED with my name here

#use add_subplot( ): add_subplot(row, column, location)
axes1 = fig.add_subplot(2, 2, 1)    #top left
axes2 = fig.add_subplot(2, 2, 2)    #top right
axes3 = fig.add_subplot(2, 2, 3)    #bottom left
axes4 = fig.add_subplot(2, 2, 4)    #bottom lefright

#add a plot to each axes: plot (x_value, y_value, format)
axes1.plot(dataset_1['x'], dataset_1['y'], 'o')
axes2.plot(dataset_2['x'], dataset_2['y'], 'o')
axes3.plot(dataset_3['x'], dataset_3['y'], 'o')
axes4.plot(dataset_4['x'], dataset_4['y'], 'o')

#add axes title
axes1.set_title('dataset_1')
axes2.set_title('dataset_2')
axes3.set_title('dataset_3')
axes4.set_title('dataset_4')

fig.tight_layout()

#The result compares the four data subset


#What is add_subplot()? 
#(row, column, location)

    # •ROW = number of rows
    # •COLUMN = number of columns
    # •POSITION/LOCATION = position/location of the graph you are plotting

    
#add_subplot(x, y, z):
    #x = rows
    #y = columns
    #z = location/position



#For example, "111" means "1x1 grid, first subplot" 
#"234" means "2x3 grid, 4th subplot".

#here, it is a 2x2 grid
```

```python
#PART III: Histograms, Scatterplot, and Regression Plot Using seaborn


#Histograms:
#Used to summarize discrete or continuous data that are measured on an interval scale. 
#It is often used to illustrate the major features of the distribution of the data in a convenient form.
#A histogram is the most commonly used graph to show frequency distributions. It looks very much like a bar chart, but different:
    #Histograms are used to show distributions of variables while bar charts are used to compare variables. 
    #Histograms visualize quantitative data or numerical data, whereas bar charts display categorical variables.


#Scatterplots:
#Use a scatter plot when you have two variables that pair well together. 
#If you have two variables that pair well together, plotting them on a scatter diagram is a great way to view their relationship 
#and see if it's a positive or negative correlation

#Regression Plot:
#used to plot data and a linear regression model fit
#a statistical method that attempts to determine the strength and character of the relationship 
#think as a scatterplot with a line through it to identify trend


```

```python
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset('tips')
tips

#Load the tips Dataset if you have not already done so
#The result shows the tips data set. 
```

```python
#Use subplots() for figure and axes
snsfig, axs = plt.subplots(1, 3, figsize=(15,5))
snsfig.suptitle('Made by Riyan Rattan')   #REPLACED with my name

#Histogram of Total Bill
axs0 = sns.histplot(tips['total_bill'], bins=10, ax=axs[0])     #make histogram
axs0.set_title('Histogram of Total Bill')

#Scatterplot of Total Bill vs Tip
axs1 = sns.scatterplot(x='total_bill', y='tip', data=tips, ax=axs[1], hue='time')   #make scatterplot
axs1.set_title('Scatterplot of Total Bill vs Tip by Time')

#regression plot of Total Bill vs Tip
axs2 = sns.regplot(x='total_bill', y='tip', data=tips, ax=axs[2])    #make regression plot
axs2.set_title('Regression of Total Bill vs Tip')
snsfig.tight_layout()
                      
# The result compares the four data subset
```

```python
#lmplot() combines regplot() and FacetGrid (helps in visualizing like colors)
#like regression but with certain areas colored across lines
```
