```python
import matplotlib.pyplot as plt
import seaborn as sns
anscombe = sns.load_dataset('anscombe')
anscombe
                                                #The result shows the anscombe data set
                                                #Import Plotting Library: matplotlib and seaborn
```

```python
dataset_1 = anscombe[anscombe['dataset']=='I']
dataset_2 = anscombe[anscombe['dataset']=='II']
dataset_3 = anscombe[anscombe['dataset']=='III']
dataset_4 = anscombe[anscombe['dataset']=='IV']
#This subset the anscombe data set. 
#(The anscombe is intentionally designed so that each data set has the same count, mean, and standard deviation.)

#subset contains set 1-4 of anscombe. Each has same count, mean, and std.
```

```python
#use figure( ) and add_subplot( ) for figure and axes
fig = plt.figure(figsize=(10,10))

#add figure title
fig.suptitle('Anscombe Data by Riyan Rattan')      #REPLACED with my name here

#use add_subplot( ): add_subplot(row, column, location)
axes1 = fig.add_subplot(2, 2, 1)    #top left
axes2 = fig.add_subplot(2, 2, 2)    #top right
axes3 = fig.add_subplot(2, 2, 3)    #bottom left
axes4 = fig.add_subplot(2, 2, 4)    #bottom lefright

#add a plot to each axes: plot (x_value, y_value, format)
axes1.plot(dataset_1['x'], dataset_1['y'], 'o')
axes2.plot(dataset_2['x'], dataset_2['y'], 'o')
axes3.plot(dataset_3['x'], dataset_3['y'], 'o')
axes4.plot(dataset_4['x'], dataset_4['y'], 'o')

#add axes title
axes1.set_title('dataset_1')
axes2.set_title('dataset_2')
axes3.set_title('dataset_3')
axes4.set_title('dataset_4')

fig.tight_layout()

#The result compares the four data subset
#Make a Figure with Axes

#What is add_subplot()? 
#(row, column, location)

#For example, "111" means "1x1 grid, first subplot" 
#"234" means "2x3 grid, 4th subplot".

#here, it is a 2x2 grid
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset('tips')
tips

#Load the tips Dataset
```

```python
#Use subplots() for figure and axes
snsfig, axs = plt.subplots(1, 3, figsize=(15,5))
snsfig.suptitle('Made by Riyan Rattan')   #REPLACED with my name

#Histogram of Total Bill
axs0 = sns.histplot(tips['total_bill'], bins=10, ax=axs[0])     #make histogram
axs0.set_title('Histogram of Total Bill')

#Scatterplot of Total Bill vs Tip
axs1 = sns.scatterplot(x='total_bill', y='tip', data=tips, ax=axs[1], hue='time')   #make scatterplot
axs1.set_title('Scatterplot of Total Bill vs Tip by Time')

#regression plot of Total Bill vs Tip
axs2 = sns.regplot(x='total_bill', y='tip', data=tips, ax=axs[2])    #make regression plot
axs2.set_title('Regression of Total Bill vs Tip')
snsfig.tight_layout()
                      
# The result compares the four data subset
```
